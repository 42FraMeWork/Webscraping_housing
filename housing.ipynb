```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
p = [f"http://www.pyclass.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/t=0&s={str(page*10)}.html"for page in range(3)]
l = [None] * len(p)
for i in range(3):

    r = requests.get(p[i], headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
    soup = BeautifulSoup(r.content, 'html.parser')
    l[i] = soup.find_all('div', {'class': 'propertyRow'})
```

```python
def num_extract(string):
    if string is None:
        return None
    else:
        return int("".join(filter(str.isdigit, string)))

def dict_prep(obj):
    if obj is None:
        return None
    else:
         return obj.text
```

```python
data = [None] * len(l)
for j, all in enumerate(l):
    values = [None] * len(all)
    for i, item in enumerate(all):
        d= {}
        
        d["Price"] = num_extract( dict_prep( item.find('h4', {'class': 'propPrice'}) ) )
        
        d["Address"] = dict_prep( item.find_all('span', {'class': 'propAddressCollapse'})[0] )
        
        d["Locality"] = dict_prep( item.find_all('span', {'class': 'propAddressCollapse'})[1] )

        d["Area"] = num_extract( dict_prep( item.find('span', {'class': 'infoSqFt'}) ) )

        d['Beds'] = dict_prep( item.find('span', {'class': 'infoBed'}) )

        d['Full Baths'] = num_extract( dict_prep(  item.find('span', {'class' : "infoValueFullBath"}) ) )

        d['Half Baths'] = num_extract ( dict_prep( item.find('span', {'class' : "infoValueHalfBath"})) )

        # TODO lot size
        # d['Lot Size'] = 


        values[i] = d
    data[j] = values
data
```

```python
df = pd.concat([pd.DataFrame(values) for values in data], ignore_index=True)
df
```
